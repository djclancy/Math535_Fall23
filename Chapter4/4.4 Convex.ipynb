**Def:** A set $D\subset\mathbb{R}^d$ is *convex* if for all $\mathbf{x},\mathbf{y}\in D$ and all $t\in[0,1]$ the point
$$
(1-t) \mathbf{x} + t \mathbf{y}\in D.
$$

In words, this means that if $\mathbf{x}$ and $\mathbf{y}$ are in a convex set $D$ then the entire line connecting $\mathbf{x}$ to $\mathbf{y}$ is also in $D$.

**Examples:**

(1) Stochastic matrix $P\in \mathbb{R}^{n\times n}$ are a convex set in $\mathbb{R}^d$;

(2) The collection $\{A\in \mathbb{R}^{n\times n}: A\text{ is positive semi-definite}\}$ is convex;

(3) Any linear subspace $U\subset \mathbb{R}^d$

**Lemma (Operations which preseverve convexity):**

Let $S_1,S_2\subset \mathbb{R}^d$, $S_3\subset\mathbb{R}^f$ be convex sets and $U\subset\mathbb{R}^d$ be a linear subspace. Let $\beta\in \mathbb{R}$ and $\mathbf{b}\in \mathbb{R}^d$. Then the following sets are convex:

(a) Scaling: $\beta S_1 = \{\beta\mathbf{x}: \mathbf{x}\in S_1\}$;

(b) Translation: $S_1 + \mathbf{b} = \{\mathbf{x}+\mathbf{b}: \mathbf{x}\in S_1\}$;

(c) Sum: $S_1+S_2 = \{\mathbf{x}_1+\mathbf{x}_2: \mathbf{x}_j\in S_j\}$;

(d) Products: ${S}_1\times S_3: \left\{\begin{bmatrix} \mathbf{x}_1 \\\mathbf{x}_3\end{bmatrix} : \mathbf{x}_j \in S_j\right\}$;

(e) Projections: $T = \{\text{proj}_U(\mathbf{x}): \mathbf{x}\in S_1\}$;

(f) Intersection: $S_1\cap S_2$.

**Def:** A function $f:\mathbb{R}^d\to \mathbb{R}$ is convex if, for all $\mathbf{x},\mathbf{y}\in \mathbb{R}^d$ and for all $t\in[0,1]$
$$
f((1-t)\mathbf{x}+t \mathbf{y}) \le (1-t) f(\mathbf{x}) + t f(\mathbf{y}).
$$

For a pictorial representation, look at the following example:

```python
def f(x):
    return np.exp(-x)+(x**2)

t_1 = np.arange(-1,2, .01)
plt.figure(figsize = (10,5))
plt.plot(t_1, f(t_1) , 'k', linewidth = 5)
plt.plot([-.87, 1.5], [f(-.87), f(1.5)], 'purple', linewidth = 5)
```

In purple is the secant line connecting points $(x_1, f(x_1))$ and $(x_2, f(x_2))$ in the $xy$-plane. In black is the curve 
$$
f(x) = e^{-x} + x^2.
$$

**Examples**

(1) Affine functions: The function $f(\mathbf{x}) = \mathbf{a}^T\mathbf{x} + b$ is convex whenever the dot-product $\mathbf{a}^T\mathbf{x}$ makes sense and $b\in \mathbb{R}$. 

(2) Distances: Both $g(\mathbf{x}) = \|\mathbf{x}\|_2$ and $f(\mathbf{x},\mathbf{y}) = \|\mathbf{x}-\mathbf{y}\|_2$ are convex.

(3) If $f:\mathbb{R}^{d+k}\to \mathbb{R}$ is convex and $C\subset \mathbb{R}^k$ is convex then
$$
g(\mathbf{x}) = \inf_{\mathbf{y}\in C} f(\mathbf{x},\mathbf{y})\qquad\text{ is convex}
$$ provided that $g(\mathbf{x})>-\infty$ for all $\mathbf{x}\in \mathbb{R}^d$. In words, if we have a function $f$ of $d+k$ variables, then taking the minimum over the last $k$ many variables (which lie in a convex set $C$) makes a convex function of the first $d$ many variables.

(4) Distance to convex sets: If $C\subset\mathbb{R}^d$ is convex then
$$
g(\mathbf{x}) = \inf_{\mathbf{y}\in C} \|\mathbf{x}-\mathbf{y}\|_2\qquad\text{ is convex}.
$$

***
Note: The value $\inf_{y\in A} f(y)$ means *infinmum* of $f(y)$ for the $y$ in $A$. It is almost exactly like the $\min_{y\in A} f(y)$ except the 
minimum operation assumes that it there is a point $y_*\in A$ such that $f(y) = \min_{y\in A} f(y)$. Formally,
$$
\inf_{y\in A} f(y) = L
$$
means (i) $f(y)\ge L$ for all $y\in A$ and (ii) that for all $\varepsilon>0$ there exists a $y$ (depending on $\varepsilon$) such that $f(y) \le L + \varepsilon$.
***

*Proof:*

(1) $$f((1-t)\mathbf{x} + t \mathbf{y} ) = \mathbf{a}^T \left((1-t) \mathbf{x} + t \mathbf{y}\right)+b
$$
$$
=\mathbf{a}^T (1-t) \mathbf{x} + \mathbf{a}^T t \mathbf{y} + b
$$
$$
= \mathbf{a}^T (1-t) \mathbf{x} + \mathbf{a}^T t \mathbf{y} + (1-t+t)b
$$
$$
=(1-t)f(\mathbf{x}) + t f(\mathbf{y}).
$$

(2) This is the triangle inequality:
$$
\|\mathbf{a}+\mathbf{b}\|_2 \le \|\mathbf{a}\|_2+\|\mathbf{b}\|_2
$$ and homogeneity
$$
\|t\mathbf{a}\|_2 = |t| \cdot \|\mathbf{a}\|_2.
$$
We only prove that $f$ is convex (the $g$ is almost the same, or apply (3) with the convex set $C= \{\mathbf{0}\}$):
$$
f( (1-t)\mathbf{x}_1 + t \mathbf{x}_2, (1-t)\mathbf{y}_1 + t \mathbf{y}_2) = \Bigg\| \Big((1-t)\mathbf{x}_1 + t \mathbf{x}_2\Big)-\Big( (1-t)\mathbf{y}_1 + t \mathbf{y}_2 \Big) \Bigg\|
$$
$$
= \Bigg\| \Big((1-t)\mathbf{x}_1 -(1-t)\mathbf{y}_1 \Big)+\Big( t \mathbf{x}_2 - t \mathbf{y}_2 \Big) \Bigg\|
$$
$$
\le\Bigg\| (1-t)\mathbf{x}_1 -(1-t)\mathbf{y}_1\Bigg\| +\Bigg\| t \mathbf{x}_2 - t \mathbf{y}_2 \Bigg\|
$$
$$
=(1-t)\Bigg\| \mathbf{x}_1 -\mathbf{y}_1\Bigg\| + t\Bigg\|  \mathbf{x}_2 -  \mathbf{y}_2 \Bigg\|
$$
$$
= (1-t) f(\mathbf{x}_1,\mathbf{y}_1) + t f(\mathbf{x}_2,\mathbf{y}_2).
$$

(3) Fix $\mathbf{x}_1,\mathbf{x}_2\in \mathbb{R}^d$ and $t\in[0,1]$. For any $\varepsilon>0$ and $j = 1,2$ there exsists some $\mathbf{y}_j\in C$ such that $$g(\mathbf{x}_j) \le f(\mathbf{x}_j, \mathbf{y}_j) \le g(\mathbf{x}_j) + \varepsilon.$$

But then
$$
g((1-t) \mathbf{x}_1 + t \mathbf{x}_2) \le f\Big((1-t) \mathbf{x}_1 + t \mathbf{x}_2, (1-t) \mathbf{y}_1 + t \mathbf{y}_2 \Big)\qquad (\text{definition of }\inf)
$$
$$
\le (1-t) f(\mathbf{x}_1, \mathbf{y}_1) + t f(\mathbf{x}_2,\mathbf{y}_2) \qquad (f \text{ is convex})
$$
$$
\le (1-t)\Big( g(\mathbf{x}_1) + \varepsilon \Big) + t \Big( g(\mathbf{x}_2) + \varepsilon\Big) \qquad (\text{ choice of }\mathbf{y}_j)
$$
$$
= (1-t) g(\mathbf{x}_1) + t g(\mathbf{x}_2) + \varepsilon.
$$
If $g((1-t) \mathbf{x}_1 + t \mathbf{x}_2) > (1-t) g(\mathbf{x}_1) + t g(\mathbf{x}_2)$ then the above **could not** hold for some $\varepsilon>0$. Thus $g((1-t) \mathbf{x}_1 + t \mathbf{x}_2) \le (1-t) g(\mathbf{x}_1) + t g(\mathbf{x}_2)$.

(4) Apply (2)+(3). $\square$

**Lemma (First order convexity):** Suppose that $f$ is continuously differentiable. Then $f$ is convex if and only if:
$$
f(\mathbf{y}) \ge f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y}-\mathbf{x}).
$$

***
*Proof:* 

Suppose that $f(\mathbf{z}_2) \ge f(\mathbf{z}_1)^T (\mathbf{z}_2 - \mathbf{z}_1)$ for all $\mathbf{z}_1,\mathbf{z}_2$. Let $\mathbf{x}, \mathbf{y}\in \mathbb{R}^d$ and $\mathbf{w} = (1-t) \mathbf{x} + t \mathbf{y}$. 

Then
$$
f(\mathbf{x}) \ge f(\mathbf{w}) + \nabla f(\mathbf{w})^T (\mathbf{x}-\mathbf{w}) 
$$
and
$$
f(\mathbf{y}) \ge f(\mathbf{w}) + \nabla f(\mathbf{w})^T (\mathbf{y}-\mathbf{w}).
$$

So for $t\in[0,1]$ we have
$$
(1-t)f(\mathbf{x})+t f(\mathbf{y}) \ge (1-t)\Big( f(\mathbf{w}) + \nabla f(\mathbf{w})^T (\mathbf{x}-\mathbf{w}) \Big) + t \Big(f(\mathbf{w}) + \nabla f(\mathbf{w})^T (\mathbf{y}-\mathbf{w})\Big)
$$
$$
= f(\mathbf{w}) + \nabla f(\mathbf{w})^T \Big((1-t)\mathbf{x}+t\mathbf{y} - (1-t)\mathbf{w}- t\mathbf{w} \Big)
$$
$$
= f(\mathbf{w}) + \nabla f(\mathbf{w})^T\mathbf{0} = f(\mathbf{w}).
$$

The other direction can be proved several ways. For example, we can use the Mean Value Theorem. Let $\mathbf{x},\mathbf{y}\in \mathbb{R}^d$ and $t\in[0,1]$. Let $\mathbf{w} = (1-t)\mathbf{x} + t\mathbf{y} = \mathbf{x} + t (\mathbf{y}-\mathbf{x})$. Then for all $t\in(0,1)$ there is some $\xi_t\in(0,1)$ such that
$$
f(\mathbf{w}) = f\Big(\mathbf{x} + t (\mathbf{y}-\mathbf{x})\Big) = f(\mathbf{x}) + \nabla f \Big(\mathbf{x} + \xi_{t} t(\mathbf{y}-\mathbf{x}) \Big)^T \Big( t(\mathbf{y}-\mathbf{x})\Big)
$$
and also
$$
f(\mathbf{w})  \le (1-t) f(\mathbf{x}) + tf (\mathbf{y}).
$$

Now we combine:
$$
(1-t) f(\mathbf{x}) + tf (\mathbf{y}) \ge f(\mathbf{x}) + \nabla f \Big(\mathbf{x} + \xi_{t} t(\mathbf{y}-\mathbf{x}) \Big)^T \Big( t(\mathbf{y}-\mathbf{x})\Big)
$$
move $f(\mathbf{x})$ over:
$$
- t f(\mathbf{x}) + tf (\mathbf{y}) \ge \nabla f \Big(\mathbf{x} + \xi_{t} t(\mathbf{y}-\mathbf{x}) \Big)^T \Big( t(\mathbf{y}-\mathbf{x})\Big)
$$
divide everything by $t>0$ :
$$
   f (\mathbf{y})- f(\mathbf{x}) \ge \nabla f \Big(\mathbf{x} + \xi_{t} t(\mathbf{y}-\mathbf{x}) \Big)^T \Big(\mathbf{y}- \mathbf{x} \Big).
$$ Taking the limit as $t\downarrow 0$ and using the fact that $f$ is continuously differentiable (and so $\nabla f$ is continuous) gives the result. $\square$

Q: What about the second derivative? 

Well for the example above 
$$
f''(x) = \frac{d^2}{dx^2} (e^{-x} + x^2) = e^{-x} + 2 \ge 2.
$$
It turns out that the behaviour of the second derivative is important for convexity as well.

**Lemma (Second-order convexity condition):** Suppose that $f:\mathbb{R}^d\to\mathbb{R}$ is twice continuously differentiable. Then $f$ is convex if and only if $H_f(\mathbf{x})$ is positive semi-definite for all $\mathbf{x}\in \mathbb{R}^d$.

***
*Proof:* 

Suppose that $H_f(\mathbf{x})\succeq 0$ (is positive semidefinite) for all $\mathbf{x}$. We can use Taylor's theorem to see that
$$
f(\mathbf{y}) = f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y}-\mathbf{x}) +\frac{1}{2} \Bigg\langle \mathbf{y}-\mathbf{x}, H_f(\mathbf{x} +\xi\Big(\mathbf{y}-\mathbf{x})\Big)(\mathbf{y}-\mathbf{x}) \Bigg\rangle 
$$
$$
\ge f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y}-\mathbf{x})
$$ and we can apply the first order condition lemma to get that $f$ is convex.

For the cother direction, we suppose that $f$ is convex. Then by Taylor again, for all $\mathbf{x}$ and $\mathbf{y}$ and $t\in(0,1)$ there exists some $\xi_t \in (0,1)$ such that 
$$
f(\mathbf{x} + t\mathbf{y})  = f(\mathbf{x}) + \nabla f(\mathbf{x})^T (t\mathbf{y}) + \frac{1}{2}\langle t \mathbf{y}, H_f(\mathbf{x}+\xi_t t \mathbf{y}) (t\mathbf{y})\rangle
$$
$$
=f(\mathbf{x}) + t \nabla f(\mathbf{x})^T \mathbf{y} + \frac{t^2}{2}\langle  \mathbf{y}, H_f(\mathbf{x}+\xi_t t \mathbf{y}) \mathbf{y}\rangle
$$ But since $f$ is convex the first order condition above implies 
$$
f(\mathbf{x} + t\mathbf{y})  \ge  f(\mathbf{x}) +  t\nabla f(\mathbf{x})^T \mathbf{y}
$$
and so rearranging and dividing by $t^2/2$ we get
$$
\langle  \mathbf{y}, H_f(\mathbf{x}+\xi_t t \mathbf{y}) \mathbf{y}\rangle \ge 0
$$ and by taking the limit as $t\to 0$ (and using the twice continuous differentiability assumption for $f$) gives 
$$
H_f(\mathbf{x}) \succeq 0.
$$ $\square$

(2) $f(\mathbf{x}) =  \langle \mathbf{x}, A\mathbf{x}\rangle + \mathbf{q}^T \mathbf{x} + r$ is convex if and only if
$$
H_f(\mathbf{x}) = A+A^T \succeq 0.
$$

**Theorem (Global minimizers of convex function):** Suppose that $f:\mathbb{R}^d\to \mathbb{R}$ is convex. Then any local minimizer $\mathbf{x}$ of $f$ is also a global minizer of $f$.

*Proof:*

Suppose not! Then there is a local minimum $\mathbf{x}$ and a global minizer $\mathbf{y}$ such that $f(\mathbf{y})< f(\mathbf{x})$. 
Then by definition of convexity
$$
f\Big( \mathbf{x}+ t(\mathbf{y} - \mathbf{x})\Big)  \le (1-t) f(\mathbf{x}) + t f(\mathbf{y}) < (1-t) f(\mathbf{x}) + t f(\mathbf{x})= f(x).
$$
By taking $t$ small, we break local minimality.
$\square$

**Theorem: (First order sufficient condition for minimality of convex)** If $f$ is convex and $\mathbf{x}\in \mathbb{R}^d$ is such that $\nabla f(\mathbf{x}) = \mathbf{0}$ then $\mathbf{x}$ is a global minimizer of $f$.
***
*Proof:* If $\nabla f(\mathbf{x}) = \mathbf{0}$ then for any other $\mathbf{y}$ (by first order condition for convexity)
$$
f(\mathbf{y}) \ge f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y} - \mathbf{x}) \ge f(\mathbf{x}).
$$
$\square$


**Example**
If $f(\mathbf{x}) = \langle \mathbf{x}, A\mathbf{x}\rangle + \mathbf{q}^T \mathbf{x} + r$ where $A+A^T\succeq 0$ then ($f$ is convex and) the global minimizers are $\mathbf{x}$ which solve
$$
\mathbf{0} = \nabla f(\mathbf{x}) = (A+A^T)\mathbf{x} + \mathbf{q}.
$$

When $A$ is symmetric and positive semi-definite this becomes
$$
2A\mathbf{x} + \mathbf{q} = \mathbf{0}.
$$

Given two matrices $A, B\in \mathbb{R}^{d\times d}$, we write $A\succeq B$ if $A-B\succeq 0$; that is $A-B$ is positive semi-definite.

**Def** Suppose function $f:\mathbb{R}^d\to \mathbb{R}$ that is twice continuously differentiable and $m>0$ is some constant. We say that $f$ is $m$-strongly convex if
$$
H_f(\mathbf{x}) - m I_{d\times d} \succeq 0 \qquad \forall \mathbf{x}\in \mathbb{R}^d.
$$

Note that $ \langle \mathbf{z} , I_{d\times d} \mathbf{z}\rangle  = \|\mathbf{z}\|^2$. So we can rewrite $H_f(\mathbf{x}) \succeq mI_{d\times d}$ as
$$
\langle \mathbf{z}, H_f(\mathbf{x})\mathbf{z}\rangle \ge m\|\mathbf{z}\|^2\qquad \forall \mathbf{z},\mathbf{x}\in \mathbb{R}^d.
$$

This leads (together with Taylor's thoerem) to

**Lemma**

Let $f:\mathbb{R}^d\to \mathbb{R}$ be twice continuously differentiable. Then $f$ is $m$-strongly convex if and only if
$$
f(\mathbf{y}) \ge f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y}-\mathbf{x}) + \frac{m}{2} \|\mathbf{y}-\mathbf{x}\|^2,\qquad \forall \mathbf{x},\mathbf{y}\in \mathbb{R}^d.
$$

**Theorem (Unique minimizer):**

Let $f:\mathbb{R}^d\to \mathbb{R}$ be twice continuously differentiable. Suppose $f$ is $m$-strongly for some $m>0$ and $\mathbf{x}^*$ is such that $\nabla f(\mathbf{x}^*) = \mathbf{0}$. Then $\mathbf{x}^*$ is the *unique* minimizer of $f$.

*Proof*:
Using the previous lemma for any $\mathbf{y}$
$$f(\mathbf{y}) \ge f(\mathbf{x}^*) + \nabla f(\mathbf{x}^*)^T (\mathbf{y}-\mathbf{x}^*) + \frac{m}{2} \|\mathbf{y}-\mathbf{x}^*\|^2 = f(\mathbf{x}^*) + \frac{m}{2} \|\mathbf{y}-\mathbf{x}^*\|^2 > f(\mathbf{x}^*) $$
and the last term is strictly positive for $\mathbf{y}\neq \mathbf{x}^*$. $\square$

**Example:** Suppose that $A$ is symmetric and positive semi-definite. Then
$$
f(\mathbf{x}) = \frac{1}{2}\langle \mathbf{x}, A\mathbf{x}\rangle + \mathbf{q}^T \mathbf{x} + r
$$
is convex. 
By the spectral theorem (or Courant-Fischer) if $\lambda_1\ge\lambda_2\ge\dotsm\ge\lambda_d\ge 0$ are the eigenvalues then $H_f(\mathbf{x}) = \frac{1}{2}(A+A^T) = A$ and so 
$$
\langle \mathbf{z}, H_f(\mathbf{x})\mathbf{z}\rangle \ge \lambda_d \|\mathbf{z}\|^2.
$$ So $H_f$ is $m$-strongly convex for all $m\in (0,\lambda_d]$.

**Example:** Let $A\in \mathbb{R}^{n\times d}$ and $\mathbf{b}\in \mathbb{R}^n$. Let us consider the least squares minization problem
$$
f(\mathbf{x}) = \|A\mathbf{x} - \mathbf{b}\|^2.
$$
Recall that $$\|\mathbf{a}-\mathbf{b}\|^2=\langle \mathbf{a}- \mathbf{b}, \mathbf{a}-\mathbf{b}\rangle = \|\mathbf{a}\|^2 + \|\mathbf{b}\|^2 - 2\langle \mathbf{a}, \mathbf{b}\rangle$$
$$
f(\mathbf{x}) = \langle A \mathbf{x}, A\mathbf{x}\rangle + \|\mathbf{b}\|^2 - 2 \langle A\mathbf{x},\mathbf{b}\rangle
$$
$$
= \frac{1}{2} \langle \mathbf{x}, 2A^T A \mathbf{x}\rangle  + (-2\mathbf{b})^T \mathbf{x} + \|\mathbf{b}\|^2.
$$

So $H_f(\mathbf{x}) = 2A^T A$ and is positive semi-definite always. If $A$ has full column rank (which implies that $d\le n$) then since $$\text{null}(A^TA) = \text{null}(A)\qquad\text{by HW problem from earlier}$$ we have that the matrix $A^TA$ is invertible. Hence the eigenvalues $\lambda_1\ge\lambda_2\ge\dotsm\ge\lambda_d>0$ of $A^TA$ are non-zero (and non-negative by positive semi-definiteness). So
$$
H_f(\mathbf{x}) \succeq 2\lambda_d I_{d\times d}.
$$

In this case, we know that there is some $\mathbf{x}^*$ which is the unique (strongly convex) when 
$$
\nabla f(\mathbf{x}) = \mathbf{0}
$$
but $\nabla f (\mathbf{x}) = \frac{1}{2}( 2 A^T A + (2A^TA)^T)\mathbf{x} - 2\mathbf{b} = 2(A^TA \mathbf{x} - \mathbf{b})$.

So the unique minizer is
$$
\mathbf{x}^* \text{ solves } A^TA \mathbf{x} = \mathbf{b}.
$$

**Example** What happens if $A$ is not of full column rank, i.e. the columns of $A$ are linearly dependent? There is a *regularization* method called ridge regression or Tikhonov regularization that tries to both fit the data $A\mathbf{x} \approx \mathbf{b}$ while throwing away some choices of $\mathbf{x}$. 

More precisely, for a parameter $\lambda>0$ (which we choose), we solve minimize
$$
f(\mathbf{x}) = \|A\mathbf{x} - \mathbf{b}\|^2_2 + \lambda \|\mathbf{x}\|^2_2. 
$$ (I'm specifying the norms because when one chooses $\lambda \|\mathbf{x}\|_1$ it is a different type of regression.)

Similar computations to above give
$$
f(\mathbf{x}) = \frac{1}{2} \langle \mathbf{x}, P\mathbf{x}\rangle + \mathbf{q}^T \mathbf{x} + r
$$
for $r = \|\mathbf{b}\|^2_2$, $P = 2(A^TA + \lambda I_{d\times d})$ and $\mathbf{q} = -2A^T \mathbf{b}$. This is strongly convex whenever $\lambda+\lambda_{\text{min}}>0$ where $\lambda_{\text{min}}$ is the smallest eigenvalue of $A^TA$. 

So the minimum is $\mathbf{x}^{**}$ which solves
$$
\mathbf{x}^{**} \text{ solves }P\mathbf{x} + \mathbf{q} = \mathbf{0}
$$
or
$$
\mathbf{x}^{**} = (A^TA+\lambda I_{d\times d})^{-1} A^T \mathbf{b}.
$$ The inverse would exist provided the above assumption on $\lambda$ and the eigenvalues of $A^TA$.
