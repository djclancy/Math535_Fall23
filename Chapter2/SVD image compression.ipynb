### Singular Value Decomposition Examples

Recall that the singular value decomposition of a matrix $A \in \mathbb{R}^{n\times m}$ by writing it as
$$
A = U \Sigma V^T
$$
where $U\in \mathbb{R}^{n\times r}$, $V\in \mathbb{R}^{m\times r}$ and $\Sigma\in \mathbb{R}^{r\times r}$ is a diagonal matrix:
$$
\Sigma = \begin{bmatrix}
\sigma_1 &0 &0 &\dots&0\\
0&\sigma_2 &0&\dots&0\\
0 &0&\sigma_3 &\dots&0\\
\vdots && \ddots &&\vdots\\
0&0&\dots &0&\sigma_r
\end{bmatrix},\qquad\sigma_1\ge\sigma_2\ge\dotsm\sigma_r>0.
$$

```python
from IPython.display import Image
import PIL
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = PIL.Image.open(".\images_for_svd\EthanHawke.jpg").convert('L')
IM = PIL.Image.open(".\images_for_svd\EthanHawke.jpg")
```

```python
plt.figure(figsize = (30,30))
plt.imshow(im, cmap=plt.get_cmap('gray'))
plt.savefig(".\images_for_svd\grayscaleFull.png")
```

```python
A = np.asarray(im)
A.shape
```

```python
(U,Sigma, V) = np.linalg.svd(A,full_matrices=False)
print(U.shape)
print(V.shape)
```

```python
plt.yscale('log')
plt.plot([t for t in range(828)], Sigma)
```

```python
kRange = [1,2,10,25,50, 150]

fig, axs = plt.subplots(2, 3)
fig.set_figheight(15)
fig.set_figwidth(15)
fig.suptitle("Number of Singular Values")
for j in range(6):
    k=kRange[j]
    B = np.matmul(np.matmul(U[:,0:k], np.diag(Sigma[0:k])),V[0:k,:])
    #plt.imshow(B, cmap=plt.get_cmap('gray'))
    if j<3:
        axs[0,j].imshow(B, cmap=plt.get_cmap('gray'))
        axs[0, j].set_title(str(k))
    else:
        axs[1,j-3].imshow(B, cmap=plt.get_cmap('gray'))
        axs[1, j-3].set_title(str(k))
plt.savefig(".\images_for_svd\severalPlots.png")
#print(np.linalg.norm(A-B))
```
